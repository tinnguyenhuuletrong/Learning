```python
import numpy as np
import pandas as pd
from io import StringIO
```

```python
def md_df(md_str: str):
    return pd.read_csv(
        StringIO(md_str.replace(' ', '').replace('-','').strip().strip('|')),  # Get rid of whitespaces
        sep='|'
    ).dropna(
        axis=1,
        how='all'
    )
```

```python
employees_data = """
| id | name  | salary | departmentId |
| 1  | Joe   | 70000  | 1            |
| 2  | Jim   | 90000  | 1            |
| 3  | Henry | 80000  | 2            |
| 4  | Sam   | 60000  | 2            |
| 5  | Max   | 90000  | 1            |
"""

departments_data = """
| id | name  |
| 1  | IT    |
| 2  | Sales |
"""

employees_1 = md_df(employees_data)
departments_1 = md_df(departments_data)

print("employees")
display(employees_1)

print("departments")
display(departments_1)
```

```python
employees_data = """
| id | name | salary | departmentId |
| 1  | Joe  | 60000  | 1            |
| 4  | Max  | 60000  | 2            |
"""

departments_data = """
| id | name |
| 1  | IT   |
| 2  | HR   |
"""

employees_2 = md_df(employees_data)
departments_2 = md_df(departments_data)

print("employees")
display(employees_2)

print("departments")
display(departments_2)
```

```python
def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    max_salary = employee.groupby("departmentId").max("salary")
    # display(max_salary)
    tmp = pd.merge(employee, max_salary, how="inner", left_on=["salary","departmentId"], right_on=["salary","departmentId"])
    # display(tmp)
    tmp = pd.merge(tmp, department, how="inner", left_on="departmentId", right_on="id")
    # display(tmp)
    res = pd.DataFrame({
        "Department": tmp["name_y"],
        "Employee":tmp["name_x"],
        "Salary":tmp["salary"],
    })
    return res
```

```python
department_highest_salary(employees_1, departments_1)
```

```python
department_highest_salary(employees_2, departments_2)
```
