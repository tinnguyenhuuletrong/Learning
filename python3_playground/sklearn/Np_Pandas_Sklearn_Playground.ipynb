## Regression

```
rng = np.random.RandomState()
x = 10 * rng.rand(50)
offset = 10 + 2 * rng.randn(50)
y = -2 * x + offset
plt.scatter(x, y);
```

```
# LinearRegression
model = LinearRegression()
model.fit(x[:, np.newaxis],y)

testY = model.predict(x[:, np.newaxis])
plt.scatter(x, y);
plt.plot(x, testY, 'r-', alpha=0.8)
```

```
# PolynomialFeatures

# x => [x, x^2, x^3, ....]
X = np.array([1,2,3])
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(X[:, None])
```

```
rng = np.random.RandomState()
x = 10 * rng.rand(50)
offset = 0.2 * rng.randn(50)
y = np.sin(x) + offset
plt.scatter(x, y);
```

```
Degree = 4

# LinearRegression -> minimize the residual sum of squares
model = make_pipeline(PolynomialFeatures(Degree), LinearRegression())
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 50)
testY = model.predict(xfit[:, np.newaxis])
fig, ax = plt.subplots(2)
ax[0].scatter(x, y);
ax[0].plot(xfit, testY, 'r-', alpha=0.8)

# model['linearregression'].coef_ => [a0,a1,a2...Degree] where 
#   a0 + a1*x + a2* x^2 + ....
ax[1].plot(np.linspace(0, Degree, Degree + 1), model[1].coef_)

fig.show()
```

```
# Ridge, Lasso -> specific loss function
#  Benefit: adjacent coefficients less up and down
model = make_pipeline(PolynomialFeatures(Degree), 
                      # Lasso(alpha=0.001)
                      Ridge())
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 50)
testY = model.predict(xfit[:, np.newaxis])
fig, ax = plt.subplots(2)
ax[0].scatter(x, y);
ax[0].plot(xfit, testY, 'r-', alpha=0.8)

# model[1].coef_ => [a0,a1,a2...Degree] where 
#   a0 + a1*x + a2* x^2 + ....
ax[1].plot(np.linspace(0, Degree, Degree + 1), model[1].coef_)

fig.show()
```

## SVM

```
x, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.80)
plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='RdBu');
```

```
#kernel: ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’
```

```
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none', edgecolors='b');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```
# linear seperator
model = SVC(kernel='linear')
model.fit(x,y)

# model.support_vectors_

plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='RdBu');
plot_svc_decision_function(model)
```

```
# RBF (radial basis function)
model = SVC(kernel='rbf')
model.fit(x,y)

# model.support_vectors_

plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='RdBu');
plot_svc_decision_function(model)
```

```
# Polynomial (a + a1*x + a2*x^2 ....)
model = SVC(kernel='poly', degree=5)
model.fit(X,y)

# model.support_vectors_

plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='RdBu');
plot_svc_decision_function(model)
```

```
# sigmoid function

model = SVC(kernel='sigmoid',coef0=-3.1)
model.fit(X,y)

# model.support_vectors_

plt.scatter(x[:, 0], x[:, 1], c=y, s=50, cmap='RdBu');
plot_svc_decision_function(model)
```

## Decision Tree

```
def visualize_classifier(model, ax=None, cmap='rainbow'):
    ax = ax or plt.gca()
    
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap=cmap, clim=(y.min(), y.max()),
                           zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
```

```
x, y = make_blobs(n_samples=300, centers=4,
                  random_state=3, cluster_std=1)
plt.scatter(x[:, 0], x[:, 1], c=y, s=5, cmap='RdBu');
```

```
model = DecisionTreeClassifier()
model.fit(x,y)

plt.scatter(x[:, 0], x[:, 1], c=y, cmap='RdBu', s=5, zorder=3);
visualize_classifier(model)
```

```
# Overfit 
x, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1)

model = DecisionTreeClassifier()
model.fit(x,y)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='RdBu', s=5, zorder=3);
visualize_classifier(model)
```

```
# Overfit reduce - HOC - BaggingClassifier
# https://www.geeksforgeeks.org/ml-bagging-classifier/

model = DecisionTreeClassifier()
hocBag = BaggingClassifier(model, n_estimators=100, max_samples=0.8)

hocBag.fit(x,y)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='RdBu', s=5, zorder=3);
visualize_classifier(hocBag)
```

```
# Overfit reduce - RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(x,y)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='RdBu', s=5, zorder=3);
visualize_classifier(model)
```

## PCA

```
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(color='black',
                    arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)
```

```
rng = np.random.RandomState()
x = 10 * rng.rand(50)
offset = 10 + 2 * rng.randn(50)
y = 2 * x + offset

X = np.column_stack((x, y))
# X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1]);
```

```
# PCA reduce demision to 1
pca = PCA(n_components=1)
pca.fit(X)
print(dict(mean=pca.mean_, dir=pca.components_, len=pca.explained_variance_))

# Normalize transform
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)

plt.scatter(X[:, 0], X[:, 1], alpha=0.2);

# Inv transform to display
X_new = pca.inverse_transform(X_pca)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8, color='r');

for length, vector in zip(pca.explained_variance_, pca.components_):
  v = vector * np.sqrt(length)
  draw_vector(pca.mean_, pca.mean_ + v)
```

```
X,_ = make_blobs(n_samples=300, centers=1, cluster_std=0.4)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2);

pca = PCA(n_components=2)
pca.fit(X)
print(dict(mean=pca.mean_, dir=pca.components_, len=pca.explained_variance_))

# Normalize transform. Should same shape ( don't reduce )
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)

for length, vector in zip(pca.explained_variance_, pca.components_):
  v = vector * np.sqrt(length)
  draw_vector(pca.mean_, pca.mean_ + v)
```

```
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
    
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = colors.T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```
